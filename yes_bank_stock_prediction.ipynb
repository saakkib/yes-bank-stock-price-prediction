# Project Summary

This project focuses on predicting the stock closing prices of YES BANK using Machine Learning techniques. Stock market prediction is one of the most challenging tasks in the financial domain because stock prices are influenced by multiple economic, political, and market-related factors. The main objective of this project was to analyze historical stock market data and build machine learning models capable of forecasting future stock closing prices with high accuracy.

The dataset used in this project contained historical stock price information of YES BANK, including features such as Date, Open Price, High Price, Low Price, and Close Price. The project began with data understanding and exploratory data analysis (EDA). Various preprocessing steps were performed, including checking missing values, duplicate records, dataset information, descriptive statistics, and correlation analysis. Several visualizations such as line charts, histograms, boxplots, heatmaps, pair plots, and correlation charts were created to identify trends, distributions, and relationships among variables.

Feature engineering was performed to improve model performance and extract useful information from the dataset. New features such as Price_Range, Average_Price, and Daily_Return were created. The Date column was transformed into Year and Month features to capture long-term and seasonal stock market trends. Highly correlated variables were removed to reduce multicollinearity and minimize overfitting risk.

The dataset was then scaled using StandardScaler to normalize feature values and improve machine learning performance. Principal Component Analysis (PCA) was also applied as an optional dimensionality reduction technique to preserve important variance while reducing feature complexity.

Multiple machine learning regression models were implemented and compared. These included:
- Linear Regression / Ridge Regression
- Decision Tree Regressor
- Random Forest Regressor

The dataset was divided into training and testing sets using an 80:20 split ratio to ensure reliable model evaluation. Model performance was evaluated using important regression metrics such as:
- MAE (Mean Absolute Error)
- MSE (Mean Squared Error)
- RMSE (Root Mean Squared Error)
- R2 Score

To improve prediction accuracy and model stability, hyperparameter optimization was performed using GridSearchCV with cross-validation. Parameters such as alpha, max_depth, n_estimators, min_samples_split, and min_samples_leaf were optimized for better performance and reduced overfitting.

Among all implemented models, Random Forest Regressor achieved the best performance with the highest R2 Score and lowest prediction error values. Feature importance analysis showed that Average_Price, Open Price, Price_Range, and Daily_Return were the most influential variables affecting stock price prediction.

Finally, the best-performing machine learning model was saved using the Joblib library for future deployment and prediction purposes. The saved model was reloaded successfully and tested on unseen data to verify prediction capability.

Overall, this project successfully demonstrated the use of machine learning techniques for stock market prediction and showed how data preprocessing, feature engineering, model optimization, and evaluation can improve predictive performance and support better financial decision-making.

````python
# Making Dataset Analysis Ready

# Create copy of dataset
df_copy = df.copy()

# Convert Date column properly
df_copy['Date'] = pd.to_datetime(df_copy['Date'], format='%b-%y')

# Extract Year and Month
df_copy['Year'] = df_copy['Date'].dt.year
df_copy['Month'] = df_copy['Date'].dt.month

# Check missing values
print("Missing Values:\n")
print(df_copy.isnull().sum())

# Remove duplicates
df_copy.drop_duplicates(inplace=True)

# Dataset info
print("\nDataset Information:\n")
df_copy.info()

# Preview dataset
print("\nProcessed Dataset Preview:\n")
display(df_copy.head())
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Handling Outliers & Outlier Treatment

# Display boxplots before treatment
plt.figure(figsize=(12,6))

sns.boxplot(data=df_copy[['Open', 'High', 'Low', 'Close']])

plt.title('Boxplot Before Outlier Treatment')

plt.show()

# ---------------------------------------------
# Outlier Treatment using IQR Method
# ---------------------------------------------

columns = ['Open', 'High', 'Low', 'Close']

for column in columns:

    # Calculate Q1 and Q3
    Q1 = df_copy[column].quantile(0.25)
    Q3 = df_copy[column].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define lower and upper limits
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Remove outliers
    df_copy = df_copy[
        (df_copy[column] >= lower_limit) &
        (df_copy[column] <= upper_limit)
    ]

# ---------------------------------------------
# Boxplot After Outlier Treatment
# ---------------------------------------------

plt.figure(figsize=(12,6))

sns.boxplot(data=df_copy[['Open', 'High', 'Low', 'Close']])

plt.title('Boxplot After Outlier Treatment')

plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model
# ML Model - 3 Implementation with Hyperparameter Optimization
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# -------------------------------------------------
# Create Base Model
# -------------------------------------------------

rf_model = RandomForestRegressor(random_state=42)

# -------------------------------------------------
# Define Hyperparameter Grid
# -------------------------------------------------

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# -------------------------------------------------
# Apply GridSearchCV
# -------------------------------------------------

grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

# -------------------------------------------------
# Fit the Algorithm
# -------------------------------------------------

grid_search.fit(X_train, y_train)

# -------------------------------------------------
# Best Parameters
# -------------------------------------------------

print("Best Parameters :")
print(grid_search.best_params_)

# -------------------------------------------------
# Best Model
# -------------------------------------------------

best_rf_model = grid_search.best_estimator_

# -------------------------------------------------
# Predict on the Model
# -------------------------------------------------

y_pred_rf = best_rf_model.predict(X_test)

# -------------------------------------------------
# Evaluation Metrics
# -------------------------------------------------

mae = mean_absolute_error(y_test, y_pred_rf)

mse = mean_squared_error(y_test, y_pred_rf)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred_rf)

# -------------------------------------------------
# Display Results
# -------------------------------------------------

print("\nEvaluation Metrics:\n")

print("MAE :", mae)
print("MSE :", mse)
print("RMSE :", rmse)
print("R2 Score :", r2)
````
